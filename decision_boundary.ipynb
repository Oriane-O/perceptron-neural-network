Visualizing boundaries

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(W, X, y, title="Decision Boundary"):
    """
    W : weights[w1, w2, b]
    X : inputs (N x 2)
    y : labels 
    """

    w1, w2, b = W

    # Decision function : w1*x1 + w2*x2 + b = 0
    # => x2 = -(w1*x1 + b) / w2
    def boundary(x):
        return -(w1 * x + b) / w2

    # Plot points
    plt.figure(figsize=(6,6))
    for xi, yi in zip(X, y):
        color = "green" if yi == 1 else "red"
        plt.scatter(xi[0], xi[1], c=color, s=120)

    # Plot line
    x_vals = np.linspace(-0.2, 1.2, 100)
    y_vals = boundary(x_vals)
    plt.plot(x_vals, y_vals, 'orange', label="Decision boundary")

    plt.xlim(-0.2, 1.2)
    plt.ylim(-0.2, 1.2)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.show()
```

FOR THE AND

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

W = np.array([0.10907717,  0.35662388, -0.39029153])  # final weights from AND training

plot_decision_boundary(W, X, y, title="Perceptron - AND")
```

FOR THE OR

```python
y = np.array([0,1,1,1])
W = np.array([0.12200567,  0.99800013, -0.0311272]) # Final weights from OR training

plot_decision_boundary(W, X, y, title="Perceptron - OR")
```

FOR XOR

```python
y = np.array([0,1,1,0])
W = np.array([0.1, -0.2, 0.05])  # example weights for XOR (never works)

plot_decision_boundary(W, X, y, title="Perceptron - XOR (fails)")
```
